# 使用pycaret训练

## 读取原始数据

```python
import pandas as pd
import numpy as np
```

```python
origin_data = pd.read_csv('GSE20194_merge.csv')
origin_data
```

```python
train_size = 0.9
fold = 8
```

```python
from pycaret.classification import ClassificationExperiment
s = ClassificationExperiment()
s.set_config('seed', 42)
s.setup(origin_data,
        target = 'group', 
        session_id = 42, 
        train_size=train_size,
        fold_strategy='kfold',
        fold=fold,
       )
```

```python
s.compare_models()
```

## 使用gan生成的平衡数据

```python
gan_data = pd.read_csv('gan_GSE20194.csv')
gan_data
```

```python
s2 = ClassificationExperiment()
s2.set_config('seed', 42)
s2.setup(gan_data, 
         target = 'group', 
         session_id = 42, 
         train_size=train_size, 
         fold_strategy='kfold',
         fold=fold,
        )
```
