# 1. 数据预处理

```python
import pandas as pd
```

```python
def GSEdata_handle(file, group):
    GSE = pd.read_csv(file)
    GSE.index = GSE['Unnamed: 0'].tolist()
    GSE = GSE.drop(['Unnamed: 0'], axis=1)
    GSE = GSE.T
    GSE['group'] = group
    return GSE
```

```python
Liver_N = GSEdata_handle('GSE20194_pCR.csv', 0)
Liver_N
```

```python
Liver_T = GSEdata_handle('GSE20194_RD.csv', 1)
Liver_T
```

# 2. 合并数据

```python
Liver = pd.concat([Liver_N, Liver_T]).dropna(axis=1)
Liver
```

```python
Liver.to_csv('GSE20194_merge.csv', index=False)
```
